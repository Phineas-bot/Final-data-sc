# Exploratory Data Analysis (EDA)

This notebook reproduces key charts from the EDA pipeline.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style='whitegrid')

base_dir = Path.cwd().parent
features_path = base_dir / 'data' / 'processed' / 'user_features.csv'
transactions_path = base_dir / 'data' / 'processed' / 'transactions_cleaned.csv'

features = pd.read_csv(features_path)
transactions = pd.read_csv(transactions_path, parse_dates=['datetime'])

display(features.head())
display(transactions.head())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=features, x='activity_level', order=['low', 'medium', 'high'])
plt.title('Distribution of Activity Levels')
plt.xlabel('Activity Level')
plt.ylabel('User Count')
plt.show()
```

```python
plt.figure(figsize=(7, 4))
order = transactions['transaction_type'].value_counts().index
sns.countplot(data=transactions, y='transaction_type', order=order)
plt.title('Transactions by Type')
plt.xlabel('Count')
plt.ylabel('Transaction Type')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(
    data=features,
    x='transactions_per_month',
    y='total_transaction_volume',
    hue='activity_level'
)
plt.title('Volume vs Frequency')
plt.xlabel('Transactions per Month')
plt.ylabel('Total Volume (XAF)')
plt.show()
```

```python
numeric_cols = features.select_dtypes(include='number')
corr = numeric_cols.corr().round(2)
plt.figure(figsize=(8, 6))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Feature Correlation Heatmap')
plt.show()
```
